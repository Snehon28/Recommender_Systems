**Extracting genres from the list of dictionaries**

```python
import ast
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
ast.literal_eval("[{'id': 16, 'name': 'Animation'}, {'id': 35, 'name': 'Comedy'}, {'id': 10751, 'name': 'Family'}]")
```

```python
df['genres'] = df['genres'].apply(convert)
```

```python
df.head(1)
```

**Making all the values to lower case of 'tags' column**

```python
def lowercase_tags(x):
    if isinstance(x, str):
        return x.lower()
    else:
        return x
```

```python
df['tags'] = df['tags'].apply(lowercase_tags)
```

**Converting tags into vectors**

```python
#Basically we are putting words into numbers
#We are extracting 5000 words from all the description concatenated together and
#we are excluding the stop_words(i.e. 'a','is','from','are','to',etc.)

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

**Removing similar looking words from the features using Natural Language Toolkit**

```python
import nltk
```

```python
#The PorterStemmer do the thing which we want(i.e. it truncate things)
##['loving','loved','love'] to ['love','love','love']

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
df['tags'][0]
```

```python
df['tags'] = df['tags'].apply(stem)
```

**Again repeating the steps**

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

**Looking for similar vectors based on angles instead of distances**

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
sorted(list(enumerate(similarity[0])),reverse = True,key = lambda x:x[1])[1:6]
```

```python
similarity[1]
```

**Finally defining a function that gives the movie recommendation based on what movie we are giving as input**

```python
def recommend(movie):
    movie_index = df[df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True, key = lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(df.iloc[i[0]].title)
```

```python
recommend('Batman Forever')
```
